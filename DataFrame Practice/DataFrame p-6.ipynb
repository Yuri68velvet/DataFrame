```python
import numpy as np
import pandas as pd
from pandas import DataFrame
```

```python
df1=pd.DataFrame([[1,2,3],[4,5,6],[7,8,9]],
                columns=['aa','bb','cc'],
                index=['AA','BB','CC'])
```

```python
df1
```

```python
df2=pd.DataFrame([[100,200,300],[400,500,600],[700,800,900]],
                 columns=['dd','ee','ff'],
                 index=['AA','BB','CC'])
```

```python
df2
```

```python
pd.concat([df1,df2],axis=1)
```

```python
pd.merge(df1,df2,right_index=True,left_index=True)
```

```python
df1.join([df2])
```

```python
df3=pd.DataFrame([[1,2,'x'],[3,4,'y'],[5,6,'z']],
    columns=['A','B','Cord1'],
    index=['1A','1B','1C'])
df3
```

```python
df4=pd.DataFrame([[100,200,'x'],[300,400,'y'],[500,600,'z']],
                columns=['C','D','Cord2'],
                index=['2A','2B','2C'])
df4
```

```python
pd.merge(df3,df4,left_on='Cord1',right_on='Cord2')
```

```python
df5=pd.DataFrame([[1000,'AA'],[4000,'BB'],[5000,'CC']],
                columns=['dd','KEY'],
                index=['x','y','z'])
df5
```

```python
df1
```

```python
pd.merge(df1,df5,left_index=True,right_on='KEY')
```

```python
df6=pd.DataFrame([[100,200],[200,300],[300,400]],
                columns=['dd','ee'],
                index=['AA','EE','FF'])
df6
```

```python
pd.concat([df1,df6],axis=1,join='inner')
```

```python
pd.concat([df1,df6],axis=1,join='outer')
```

```python
pd.concat([df1,df6],axis=1,join_axes=[df1.index])
```

```python
pd.concat([df1,df6],axis=1,join_axes=[df6.index])
```

```python
pd.merge(df1,df2,left_index=True,right_index=True,how='inner')
```

```python
pd.merge(df1,df2,left_index=True,right_index=True,how='outer')
```

```python
pd.merge(df1,df2,left_index=True,right_index=True,how='left')
```

```python
pd.merge(df1,df2,left_index=True,right_index=True,how='right')
```
