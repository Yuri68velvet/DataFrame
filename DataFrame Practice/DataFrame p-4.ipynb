```python
left=pd.DataFrame({'No':[1,2,3,4],
                  'Name':['安川','井上','中村','野間'],
                  'Level':['BS','BT','BT','BG']})
right=pd.DataFrame({'No':[1,2,3,4],
                   'Name':['執行','長浜','不動寺','中村'],
                  'Level':['SM','SSV1','SSV2','ASM']})
```

```python
left
```

```python
right
```

```python
pd.merge(left,right,how='inner',on='No')
```

```python
left=pd.DataFrame({'No1':[1,2,3,4],
                   'PTR':['B','B','B','B'],
                  'Name':['安川','井上','中村','野間'],
                  'Level':['BS','BT','BT','BG']})
right=pd.DataFrame({'No1':[1,2,3,4],
                    'PTR':['SSV','B','B','SSV'],
                   'Name':['執行','長浜','不動寺','中村'],
                  'Level':['SM','SSV1','SSV2','ASM']})
```

```python
pd.merge(left,right,on=['No1','PTR'])
```

```python
pd.merge(left,right,how='outer',on=['No1','PTR'])
```

```python
pd.merge(left,right,how='right',on=['No1','PTR'])
```

```python
pd.merge(left,right,how='left',on=['No1','PTR'])
```

```python
left=pd.DataFrame({'A':['安川','松村','今村','世継'],
                  'B':['中原','篠崎','松田','大倉']},
                   index=['CLA','RES','CLA','RES'])
right=pd.DataFrame({'C':['阿部','森山','菊永','河野'],
                   'D':['迫','笹尾','柴崎','立野']},
                  index=['CLA','RES','CLA','RES'])
```
